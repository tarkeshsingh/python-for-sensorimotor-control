## 4. Language Semantics — Indentation & Whitespace <a id='4.-Language-Semantics'></a>

Unlike many other languages (C++, Java, R) that use **braces `{}`** to define code blocks, Python uses **whitespace (indentation)**. A colon (`:`) marks the start of an indented block, and all lines within that block must be indented by the same amount (typically 4 spaces).

This makes Python code visually clean, but you must be careful with indentation — it's not just style, it's **syntax**.

Let's see this in action with a sorting algorithm where indentation defines which statements belong to which loop:

````python
# ---- Demonstrating indentation with Bubble Sort ----
# Notice how indentation level determines which block each line belongs to.

numbers = [5, 2, 9, 1, 5, 6]
print("Unsorted list:", numbers)

# Bubble Sort Algorithm
for i in range(len(numbers)):                          # OUTER loop (level 1)
    for j in range(0, len(numbers) - i - 1):           # INNER loop (level 2)

        if numbers[j] > numbers[j + 1]:                # Condition (level 3)
            # Swap adjacent elements if they're in the wrong order
            numbers[j], numbers[j + 1] = numbers[j + 1], numbers[j]

            # This print is inside the 'if' block (level 4)
            print(f"Swapped {numbers[j + 1]} and {numbers[j]} → {numbers}")

    # This print is inside the OUTER loop but outside the INNER loop (level 1)
    print(f"After pass {i + 1}: {numbers}")

# This print is OUTSIDE all loops (level 0)
print("\nSorted list:", numbers)
````

### Key rules for indentation

- A **colon** (`:`) denotes the start of an indented code block (after `if`, `for`, `while`, `def`, `class`, etc.).
- Everything in the block must be indented by the **same amount** (4 spaces is standard).
- When the indentation level decreases, you've exited the block.
- **Mixing tabs and spaces** will cause errors — stick to spaces (most editors handle this automatically).

````python
# ---- Multiple statements on one line (semicolons) ----
# You CAN separate statements with semicolons, but this is generally discouraged
# because it reduces readability.

a = 1; b = 2; c = 4  # Legal, but prefer separate lines in real code
print(a, b, c)
````

## 5. Variables & Argument Passing <a id='5.-Variables-&-Argument-Passing'></a>

### Variables are references, not boxes

When you assign a variable in Python, you create a **reference** (a pointer) to the object on the right side of the equals sign. You are **not** copying the data — you're giving it a name.

This distinction is critical when working with large datasets in movement neuroscience. If you accidentally modify a shared reference, you could corrupt your original data.

````python
# ---- Variables are references, not copies ----

a = [11, 23, 34]  # Create a list and assign it to 'a'
````

````python
# ---- Assigning b = a does NOT copy the list ----
# Both 'a' and 'b' now point to the SAME list object in memory.

b = a
print(b)  # [11, 23, 34]
````

````python
# ---- Proof: modifying 'a' also changes 'b' ----
# Because they reference the same object!

a.append(4)  # Add 4 to the list via 'a'
print(b)     # [11, 23, 34, 4] — 'b' sees the change too!
````

**This is a common source of bugs.** If you want an independent copy of a list, use `b = a.copy()` or `b = list(a)`. For NumPy arrays, use `b = a.copy()`.

### Everything is an object

An important characteristic of Python is the consistency of its **object model**. Every number, string, data structure, function, class, and module exists as a Python object. Each object has an associated **type** (e.g., `int`, `str`, `list`, `function`) and internal data. This makes the language very flexible — even functions can be stored in variables, passed as arguments, and returned from other functions.

### Passing arguments to functions

Understanding Python's reference semantics is critical when working with large datasets. When you pass an object to a function, the function receives a **reference** to that object, not a copy. This means that if the function modifies the contents of a mutable argument (like a list), the changes are visible outside the function.

````python
# ---- Functions receive references, not copies ----
# This means a function can modify the original list you pass in.

def append_element(some_list, element):
    """Append an element to the given list (modifies it in place)."""
    some_list.append(element)

data = [1, 2, 3]
append_element(data, 4)  # The function modifies 'data' directly
print(data)  # [1, 2, 3, 4] — the original list was changed!
````

**Lab analogy:** Imagine `data` is a lab notebook. If you hand someone the actual notebook (a reference), any notes they write go into your notebook. If you hand them a photocopy (a copy), your original stays untouched.

## 6. Mutable vs Immutable Objects <a id='6.-Mutable-vs-Immutable-Objects'></a>

Python objects fall into two categories:

| Mutable (can be changed) | Immutable (cannot be changed) |
|---|---|
| `list`, `dict`, `set`, NumPy arrays, most user-defined classes | `int`, `float`, `str`, `tuple`, `frozenset` |

Understanding this distinction helps you avoid unexpected bugs and write safer code.

````python
# ---- Mutable example: lists can be modified in place ----

a_list = ["foo", 2, [4, 5]]
print(f"Before: {a_list}")

a_list[2] = (3, 4)  # Replace the third element
print(f"After:  {a_list}")
````

````python
# ---- Immutable example: tuples CANNOT be modified ----
# This will raise a TypeError — uncomment to see the error.

a_tuple = (3, 5, (4, 5))
# a_tuple[1] = "four"  # TypeError: 'tuple' object does not support item assignment
````

````python
# ---- Immutable example: strings CANNOT be modified in place ----
# You must create a NEW string instead.

a = "this is a string"
# a[10] = "f"  # TypeError: 'str' object does not support item assignment

# To "modify" a string, use methods that return a NEW string:
b = a.replace("string", "longer string")
print(f"b = {b}")  # New string
print(f"a = {a}")  # Original is unchanged
````

**Why this matters in research:** When you load a raw EMG signal into a NumPy array (mutable), any filtering or processing you do modifies the original unless you explicitly make a copy first. A common pattern is:
```python
raw_emg = load_emg_data()       # Original data
filtered_emg = raw_emg.copy()   # Work on a copy
apply_bandpass_filter(filtered_emg)
```

## 7. Scalar Types — Numbers, Strings, Booleans, and None <a id='7.-Scalar-Types'></a>

Python has a small set of built-in types for handling numerical data, strings, Boolean (`True`/`False`) values, and the special `None` value. These "single value" types are sometimes called **scalar types**.

### 7.1 Numeric Types: `int` and `float`

````python
# ---- Integers (int) ----
# Python integers can be arbitrarily large — no overflow!

ival = 17239871
print(f"ival ** 7 = {ival ** 7}")  # ** is the exponentiation operator
````

````python
# ---- Floating-point numbers (float) ----
# Under the hood, each float is a double-precision (64-bit) value.
# You can use scientific notation with 'e'.

fval = 7.243
fval2 = 6.78e-5  # = 0.0000678 (scientific notation)

print(f"fval = {fval}, fval2 = {fval2}")
````

````python
# ---- Division behavior ----

# Regular division always returns a float:
print(f"5 / 6 = {5 / 6}")    # 0.8333...

# Floor division (//) drops the fractional part:
print(f"7 // 3 = {7 // 3}")  # 2 (integer result)
````

### 7.2 Strings

Strings are sequences of characters. You'll use them constantly for labeling trials, formatting output, building file paths, and more.

````python
# ---- Creating strings ----
# You can use single quotes or double quotes (double quotes are preferred by convention).

a = 'one way of writing a string'
b = "another way"
print(a)
````

````python
# ---- Multiline strings use triple quotes ----

c = """
This is a longer string that
spans multiple lines
"""
print(c)

# Count the newline characters — there are 3 (before, between, and after the text lines)
print(f"Number of newlines: {c.count(chr(10))}")
````

````python
# ---- Strings are IMMUTABLE ----
# You cannot modify a character in place — you must create a new string.

a = "this is a string"
# a[10] = "f"  # This would raise: TypeError: 'str' object does not support item assignment

# Instead, use .replace() which returns a NEW string:
b = a.replace("string", "longer string")
print(f"b: {b}")
print(f"a: {a}")  # Original is unchanged
````

````python
# ---- Converting objects to strings ----
# The str() function converts almost anything to a string.

a = 5.6
s = str(a)
print(f"Value: {s}, Type: {type(s)}")  # '5.6' as a string
````

````python
# ---- Strings as sequences: indexing and slicing ----
# Strings behave like lists of characters.

s = "python"
print(f"Characters: {list(s)}")  # ['p', 'y', 't', 'h', 'o', 'n']
print(f"First 3:   {s[:3]}")     # 'pyt' — slicing syntax: s[start:stop]
````

````python
# ---- String concatenation ----
# The + operator joins strings together.

a = "this is the first half "
b = "and this is the second half"
print(a + b)
````

### String Formatting

Python offers two main ways to create formatted strings. The newer **f-string** syntax (introduced in Python 3.6) is generally preferred for readability.

````python
# ---- Method 1: .format() (older style) ----
# {0:.2f} → format first argument as float with 2 decimal places
# {1:s}   → format second argument as string
# {2:d}   → format third argument as integer

template = "{0:.2f} {1:s} are worth US${2:d}"
result = template.format(88.46, "Argentine Pesos", 1)
print(result)
````

````python
# ---- Method 2: f-strings (preferred — more readable) ----
# Prefix the string with 'f' and embed expressions directly in {curly braces}.

amount = 10
rate = 88.46
currency = "Pesos"

# You can do math inside the braces and add format specifiers after a colon:
print(f"{amount} {currency} is worth US${amount / rate:.2f}")

# Movement neuroscience example:
participant = "P01"
mean_rt = 0.3127
print(f"Participant {participant}: mean reaction time = {mean_rt:.3f} s")
````

### Common String Methods

Strings have many built-in methods that you'll use constantly in research code — parsing file paths, building condition labels, cleaning data exported from sensors. Here are the most important ones:

| Method | What it does | Example |
|---|---|---|
| `.split(sep)` | Split string into a list at each `sep` | `"a,b,c".split(",")` → `['a','b','c']` |
| `.join(list)` | Join a list into one string with separator | `"_".join(['P01','ctrl'])` → `'P01_ctrl'` |
| `.strip()` | Remove leading/trailing whitespace | `"  hello  ".strip()` → `'hello'` |
| `.replace(old, new)` | Replace all occurrences | `"foo bar".replace("bar","baz")` → `'foo baz'` |
| `.upper()` / `.lower()` | Change case | `"Hello".lower()` → `'hello'` |
| `.startswith(prefix)` | Check if string starts with prefix | `"P01_ctrl".startswith("P01")` → `True` |
| `.endswith(suffix)` | Check if string ends with suffix | `"data.csv".endswith(".csv")` → `True` |

````python
# ---- .split(): break a string into a list ----
# Extremely common for parsing file paths and CSV-like strings.

path = "/content/drive/MyDrive/data/P01_session2.csv"
parts = path.split("/")           # Split at every '/'
print(f"Path parts: {parts}")
print(f"Filename:   {parts[-1]}")   # Last element = filename

# Split a condition label:
label = "P01_control_trial05"
fields = label.split("_")
print(f"\nFields: {fields}")        # ['P01', 'control', 'trial05']
````

````python
# ---- .join(): combine a list into one string ----
# The opposite of .split(). The separator goes BEFORE .join().

parts = ["P01", "perturbation", "block3"]
filename = "_".join(parts) + ".csv"     # 'P01_perturbation_block3.csv'
print(f"Built filename: {filename}")

# Join with a different separator:
print(f"Comma-separated: {', '.join(parts)}")
print(f"No separator:    {''.join(['a', 'b', 'c'])}")
````

````python
# ---- .strip(), .startswith(), .endswith() ----
# Useful for cleaning data and filtering files.

# Strip whitespace (common in sensor exports with trailing spaces/newlines):
messy = "  0.3421  \n"
clean = messy.strip()
print(f"Stripped: '{clean}'")

# Check file extensions:
filename = "P01_session2.csv"
print(f"Is CSV?  {filename.endswith('.csv')}")
print(f"Is P01?  {filename.startswith('P01')}")

# Practical: filter a list of filenames
all_files = ["P01_data.csv", "P02_data.csv", "notes.txt", "P01_backup.csv"]
csv_files = [f for f in all_files if f.endswith(".csv")]
p01_files = [f for f in all_files if f.startswith("P01")]
print(f"\nCSV files: {csv_files}")
print(f"P01 files: {p01_files}")
````

### 7.3 Booleans

The two Boolean values in Python are `True` and `False`. Comparisons and conditional expressions evaluate to one of these. Boolean values are combined with the `and`, `or`, and `not` keywords.

````python
# ---- Boolean logic ----

print(f"True and True  = {True and True}")    # True
print(f"True or False  = {True or False}")     # True
print(f"not True       = {not True}")           # False

# When converted to numbers: False → 0, True → 1
print(f"int(False)     = {int(False)}")         # 0
print(f"int(True)      = {int(True)}")          # 1

# This is useful! e.g., count how many trials were "successful":
trial_results = [True, True, False, True, False]
print(f"Successful trials: {sum(trial_results)} out of {len(trial_results)}")
````

### 7.4 Type Casting

The `str`, `bool`, `int`, and `float` types double as **functions** that convert values between types.

````python
# ---- Type casting examples ----

s = "3.14159"              # A string
fval = float(s)            # Convert string → float
print(f"float('{s}') = {fval}, type = {type(fval)}")

ival = int(fval)           # Convert float → int (truncates decimal)
print(f"int({fval}) = {ival}")

bval = bool(fval)          # Any non-zero number → True; 0 → False
print(f"bool({fval}) = {bval}")
print(f"bool(0) = {bool(0)}")
````

### 7.5 None — Python's "nothing" value

`None` is Python's null value — it represents the absence of a value. It's commonly used as a default for optional function arguments and to indicate missing data.

````python
# ---- None basics ----
# Use 'is' (not ==) to check for None.

a = None
print(f"a is None: {a is None}")       # True
print(f"a is not None: {a is not None}") # False
````

````python
# ---- None as a default function argument ----
# This is a very common pattern in Python.
# Movement neuroscience example: optionally scale a force value.

def add_and_maybe_multiply(a, b, c=None):
    """
    Add a + b, and optionally multiply by c if provided.

    Parameters
    ----------
    a, b : numbers to add
    c : optional multiplier (default: None, meaning no multiplication)
    """
    result = a + b
    if c is not None:
        result = result * c
    return result

print(f"add_and_maybe_multiply(2, 3)    = {add_and_maybe_multiply(2, 3)}")      # 5
print(f"add_and_maybe_multiply(2, 3, 4) = {add_and_maybe_multiply(2, 3, 4)}")  # 20
````

## 8. Control Flow — `if`, `for`, `while`, and Friends <a id='8.-Control-Flow'></a>

Control flow statements let your program make decisions and repeat actions. These are the workhorses of any data analysis pipeline.

### 8.1 `if`, `elif`, `else`

The `if` statement checks a condition. If `True`, it executes the indented block that follows. You can chain additional conditions with `elif` ("else if") and catch everything else with `else`.

````python
# ---- Basic if statement ----

x = -5
if x < 0:
    print("It's negative")  # Note the indent — this line is inside the 'if' block
````

````python
# ---- if / elif / else chain ----
# Only the FIRST matching condition is executed.

x = -5

if x < 0:
    print("It's negative")
elif x == 0:
    print("Equal to zero")
elif 0 < x < 5:  # Python allows chained comparisons!
    print("Positive but smaller than 5")
else:
    print("Positive and larger than or equal to 5")
````

````python
# ---- Chained comparisons ----
# Python lets you write mathematical-style chains like 1 < x < 10.

print(f"3 > 1 > 5 is {3 > 1 > 5}")  # False (because 1 > 5 is False)
print(f"1 < 3 < 5 is {1 < 3 < 5}")  # True
````

### 8.2 `for` Loops

`for` loops iterate over a collection (like a list, tuple, or range). They're the primary way to process data trial-by-trial, sample-by-sample, or participant-by-participant.

````python
# ---- for loop with 'continue' ----
# 'continue' skips the rest of the current iteration and moves to the next one.
# This is useful for skipping missing data (None values).

# Imagine these are force readings, with some missing values:
sequence = [1, 2, None, 4, None, 5]
total = 0

for value in sequence:
    if value is None:
        continue       # Skip this iteration — go to the next value
    total += value     # Only runs if value is not None

print(f"Sum (skipping None): {total}")  # 12
````

````python
# ---- for loop with 'break' ----
# 'break' exits the loop entirely.
# Example: sum values until you hit a 5 (a sentinel value).

sequence = [1, 2, 0, 4, 6, 5, 2, 1]
total_until_5 = 0

for value in sequence:
    if value == 5:
        break                # Stop the loop immediately
    total_until_5 += value

print(f"Sum until 5: {total_until_5}")  # 1+2+0+4+6 = 13
````

````python
# ---- break only exits the INNERMOST loop ----
# Outer loops continue running.

for i in range(4):
    for j in range(4):
        if j > i:
            break          # Exits the inner loop only
        print((i, j))
````

### 8.3 `while` Loops

A `while` loop repeats a block of code as long as its condition remains `True`. Use `break` to exit early.

````python
# ---- while loop with a break condition ----
# Keep halving x and accumulating the total, but stop if total exceeds 500.

x = 256
total = 0

while x > 0:
    if total > 500:
        break          # Safety valve — stop if total gets too large
    total += x
    x = x // 2         # Floor division: halve x each iteration
    print(f"x = {x}, total = {total}")
````

### 8.4 `pass` — The "Do Nothing" Statement

`pass` is a placeholder that does nothing. You use it when Python requires a statement (e.g., inside an `if` or `class` block) but you haven't written the code yet.

````python
# ---- pass as a placeholder ----

x = 128  # From the while loop above

if x < 0:
    print("negative!")
elif x == 0:
    # TODO: handle the zero case later
    pass              # Placeholder — without this, Python would raise a SyntaxError
else:
    print("positive!")
````

### 8.5 `range()` — Generating Sequences of Integers

The `range()` function generates a sequence of evenly spaced integers. It's the most common way to create loop counters.

````python
# ---- range() basics ----

# range(n) generates integers from 0 to n-1
print(f"range(10): {list(range(10))}")        # [0, 1, 2, ..., 9]

# range(start, stop, step)
print(f"range(0, 20, 2): {list(range(0, 20, 2))}")  # Even numbers: [0, 2, 4, ..., 18]
print(f"range(5, 0, -1): {list(range(5, 0, -1))}")  # Countdown: [5, 4, 3, 2, 1]
````

````python
# ---- Using range() with a for loop ----
# A common pattern for iterating with an index.

seq = [1, 2, 3, 4]
for i in range(len(seq)):
    print(f"element {i}: {seq[i]}")

# Note: In practice, enumerate() is often preferred (covered in later tutorials):
# for i, val in enumerate(seq):
#     print(f"element {i}: {val}")
````
